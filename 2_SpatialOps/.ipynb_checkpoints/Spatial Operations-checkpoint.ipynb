# Digital Image Processing: Spatial Operations

## 1. Loading modules

````python
import cv2
import ipywidgets as widgets
import numpy as np
import os

from glob import glob

from ipywidgets import interact, interactive, fixed
from IPython.display import display

from matplotlib import pyplot as plt

%matplotlib inline
````

## 2. Convolution theory

$$ g(t) = x(t) \circledast h(t) $$

````python
def convolve(img, kernel, flip=True):
    height, width = img.shape
    kernel_flip = np.flipud(np.fliplr(kernel))
    out_img = np.zeros(img.shape)

    # Zero padding.
    img_padded = np.zeros((height + 2, width + 2))
    img_padded[1:-1, 1:-1] = img
    for row in range(height):     # Loop over every pixel of the image
        for column in range(width):
            out_img[row,column]=(kernel_flip*img_padded[row:row + 3,column:column + 3]).sum()        
    return out_img
````

Testing convolution ans visualizing result.

````python
kernel = np.ones((3, 3)) / 9
img = cv2.imread('images/blurry_moon.png', cv2.IMREAD_GRAYSCALE)
out = convolve(img, kernel)

plt.figure(figsize=(7, 7))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray'); plt.axis('off')
plt.title('Original image');

plt.subplot(1, 2, 2)
plt.imshow(out, cmap='gray'); plt.axis('off')
plt.title('Convolved image (smoothing)');
````

## 3. Spatial operations using OpenCV

We will apply most convolutions using OpenCV built-in function ```cv2.Filter2D```

### Image smoothing

````python
def smooth_image(img, k_size):
    kernel = np.ones((k_size, k_size)) / k_size ** 2
    img_filter = cv2.filter2D(img, cv2.CV_8U, kernel)
    
    plt.figure(figsize=(10, 10))
    plt.subplot(1, 2, 1)
    plt.imshow(img, cmap='gray'); plt.axis('off')
    plt.title('Original image');

    plt.subplot(1, 2, 2)
    plt.imshow(img_filter, cmap='gray'); plt.axis('off')
    plt.title('Smoothed image with kernel size {0}'.format(k_size));    
    
img = cv2.imread('images/test_pattern_blurring_orig.png', cv2.IMREAD_GRAYSCALE)  
kernel_slider = widgets.IntSlider(min=3, max=50, value=9, step=2, description='kernel size')
interact(smooth_image, img=fixed(img), k_size=kernel_slider);
````

### Image sharpening (laplacian filtering)

````python
kernel = np.array([[0,1,0],[1,-4,1],[0,1,0]])
img = cv2.imread('images/blurry_moon.png', cv2.IMREAD_GRAYSCALE)
img_filter = cv2.filter2D(img, cv2.CV_32F, kernel)
img_sharpen = img - img_filter

plt.figure(figsize=(13, 13))

plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray'); plt.axis('off')
plt.title('Original image');

plt.subplot(1, 3, 2)
plt.imshow(img_filter, cmap='gray'); plt.axis('off')
plt.title('Laplacian filter');

plt.subplot(1, 3, 3)
plt.imshow(img_sharpen, cmap='gray', vmin=0, vmax=255); plt.axis('off')
plt.title('Image sharpened');
````

### Image correlation

Correlation is similar to convolution operation. The difference is that you don't need to flip the kernel. If the kernel is simmetrical, convolution and correlation are the same.

````python
img = cv2.imread('images/messi.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

template = cv2.imread('images/messi_face.jpg')
template = cv2.cvtColor(template, cv2.COLOR_BGR2RGB)
height, width, _ = template.shape

plt.figure()
plt.axis('off');plt.imshow(img); plt.title('Original image');
plt.figure()
plt.axis('off');plt.imshow(template); plt.title('Template image');
````

Applying correlation

````python
res = cv2.matchTemplate(img, template, cv2.TM_CCOEFF)
_, _, _, top_left = cv2.minMaxLoc(res)
bottom_right = (top_left[0] + width, top_left[1] + height)
cv2.rectangle(img,top_left, bottom_right, (0, 255, 0), 2)

plt.imshow(img); plt.axis('off'); plt.title('Output image');
````

````python
img = cv2.imread('images/mario.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

template = cv2.imread('images/coin.png')
template = cv2.cvtColor(template, cv2.COLOR_BGR2RGB)
height, width, _ = template.shape

res = cv2.matchTemplate(img, template, cv2.TM_CCOEFF_NORMED)
locations = np.where(res >= 0.8)
for point in zip(locations[1], locations[0]):
    cv2.rectangle(img, point, (point[0] + width, point[1] + height), (0, 255, 0), 2)

plt.imshow(img); plt.axis('off'); plt.title('Output image');
````
