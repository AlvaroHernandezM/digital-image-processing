# Instalación y configuración

Para este taller utilizaremos Python 3.5 o superiores y las siguientes librerías o módulos:

- OpenCV
- Numpy
- Matplotlib
- Jupyter

## Python

> Tomado de https://tutorial.djangogirls.org/es/python_installation/

Python es un lenguaje de programación muy popular que puede utilizarse para la creación de sitios web, juegos, software académico, gráficos y mucho, mucho más.

Python se originó en la década de 1980 y su objetivo principal es ser legible por los seres humanos (¡no sólo para las máquinas!), por eso parece mucho más simple que otros lenguajes de programación. Esto hace que sea más fácil de aprender, pero no te preocupes, ¡Python es también muy poderoso!


## Windows

### Descarga de WinPython

La manera más sencilla de instalar Python junto con los módulos mencionados anteriormente (excepto por OpenCV), es utilizando el instalador de **WinPython** https://sourceforge.net/projects/winpython/files/WinPython_3.5/3.5.3.1/ Descargar la opción Qt, dependiendo de la arquitectura de tu computadora.

### Instalación de OpenCV.

Utilizando la consola de comandos de WinPython (WinPython Command Prompt), resulta sencillo la instalación de paquetes o librerías de Python, ejecutando comandos de pip. Para instalar opencv, se debe escribir en consola la siguiente línea de comandos y ejecutar:

```sh
pip install opencv-python opencv-contrib-python
```

## Linux (Configurado en debian y similares).

### Instalación de Python3
```sh
sudo apt-get install python3 python3-virtualenv
```

### Instalación y creación de ambiente virtual
```sh
$ virtualenv --no-site-packages --python=/usr/bin/python3 dip/
$ source dip/bin/activate
```

### Instalación de los paquetes
```sh
$ pip install numpy matplotlib opencv-python opencv-contrib-python jupyter
```

## Mac

### Instalación de Homebrew.
Entrar al siguiente enlace para instrucciones http://brew.sh/index_es.html

### Instalación de Python3
```sh
$ brew install python3 pip
$ pip install virtualenv
```
### Instalación de OpenCV y librerías.
```
$ virtualenv --no-site-packages --python=/usr/local/bin/python3 dip/
$ source dip/bin/activate
$ pip install numpy matplotlib opencv-python opencv-contrib-python jupyter
```
